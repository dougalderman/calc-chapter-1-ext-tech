```r
x<-c(0,50,100,200,285,300,360,400)
y<-c(4.5,43,82,130,142,134,100,60)
```

```r
plot(x,y,col="blue", main="Baseball Home Run Trajectory",xlab="Horizontal Distance (Feet)", ylab="Vertical Distance (Feet)")
```

```r
relation3<-lm(y~x+I(x^2)+I(x^3))
print(relation3)
summary(relation3)
```

```r
x_seq <- seq(from=0,to=500, by=0.1)
new_relation3<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation3[i]<-predict(relation3, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation3),xlab="x",ylab="y")
```

```r
confint(relation3, level=0.95)
```

```r
plot(fitted(relation3), residuals(relation3))
```

```r
x_seq_pred3 <- seq(from=400,to=500, by=0.1)
y_pred3<-list(0)
for (i in 1:length(x_seq_pred3)) {
    x_pred3 <- data.frame(x=x_seq_pred3[i])
    y_pred3[i]<-predict(relation3, x_pred3)
    print(x_pred3)
    print(y_pred3[i])
}
x_seq_pred3_range <- c(400,500)
plot(x,y,xlim=x_seq_pred3_range,col="blue", main="Linear Regression Example",lines(x_seq_pred3, y_pred3),xlab="x",ylab="y")
```

```r
relation_cf<-nls(y~0.0015*x*sqrt(202500-x^2), start=list(x=0))
print(relation_cf)
summary(relation_cf)
```
